# 1. EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO

# Set plot style and default figure size
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# --- 1. Load the Data ---
try:
    #df = pd.read_csv(StringIO(csv_data))
    print("Dataset loaded successfully from sample data.")
    # To load from a file:
    file_path = 'data/HR_Assessment_Data_Data Scientist_Raw Data.csv'
    df = pd.read_csv(file_path)
    print(f"Dataset loaded successfully from {file_path}")
except FileNotFoundError:
    print("Error: The specified CSV file was not found.")
    exit()
except Exception as e:
    print(f"An error occurred while loading the data: {e}")
    exit()

# --- 2. Initial Inspection & Key Characteristics ---

print("\n--- Dataset Overview ---")
print(f"Shape (Rows, Columns): {df.shape}")
print("\n--- First 5 Rows ---")
print(df.head())
print("\n--- Data Types and Non-Null Counts ---")
print(df.info())
print("\n--- Summary Statistics for Numerical Features ---")
print(df.describe(include=[np.number]))
print("\n--- Summary Statistics for Categorical Features ---")
print(df.describe(include=['object']))

# --- 3. Data Cleaning & Issue Identification ---

print("\n--- Checking for Missing Values ---")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
if missing_values.sum() == 0:
    print("No missing values found.")
else:
    print(f"Total missing values: {missing_values.sum()}")
    # Consider imputation or dropping rows/columns if necessary

print("\n--- Checking for Duplicate Rows ---")
duplicate_rows = df.duplicated().sum()
if duplicate_rows == 0:
    print("No duplicate rows found.")
else:
    print(f"Found {duplicate_rows} duplicate rows.")
    # Optional: df = df.drop_duplicates(keep='first', inplace=False)

print("\n--- Checking Data Type Correctness & Value Ranges ---")
# Convert timestamp columns to datetime objects
try:
    df['start_time'] = pd.to_datetime(df['start_time'])
    df['end_time'] = pd.to_datetime(df['end_time'])
    print("Converted 'start_time' and 'end_time' to datetime objects.")
    time_conversion_success = True
except Exception as e:
    print(f"Error converting time columns: {e}")
    time_conversion_success = False

# Verify rating range (assuming 1-5)
if 'rating' in df.columns:
    min_rating, max_rating = df['rating'].min(), df['rating'].max()
    print(f"Rating range: {min_rating} - {max_rating}")
    if not (df['rating'].between(1, 5, inclusive='both').all()):
        print("Warning: Ratings found outside the expected 1-5 range.")

# Verify cancelled values (assuming 0 or 1)
if 'cancelled' in df.columns:
    unique_cancelled = df['cancelled'].unique()
    print(f"Unique values in 'cancelled': {unique_cancelled}")
    if not all(val in [0, 1] for val in unique_cancelled):
        print("Warning: Unexpected values found in 'cancelled' column.")

# Verify non-negative fare and distance
if 'fare' in df.columns and (df['fare'] < 0).any():
    print("Warning: Negative values found in 'fare'.")
if 'distance_km' in df.columns and (df['distance_km'] < 0).any():
    print("Warning: Negative values found in 'distance_km'.")

# Sort dataframe by start time for time-based features
if time_conversion_success:
    df = df.sort_values(by='start_time').reset_index(drop=True)
    print("DataFrame sorted by 'start_time'.")

# --- 4. Feature Engineering ---

print("\n--- Feature Engineering: Basic Time & Speed Features ---")
if time_conversion_success:
    # Calculate trip duration in minutes
    df['trip_duration_minutes'] = (df['end_time'] - df['start_time']).dt.total_seconds() / 60
    print("Created 'trip_duration_minutes' feature.")
    if (df['trip_duration_minutes'] <= 0).any():
        print("Warning: Found trips with non-positive duration.")

    # Extract time features
    df['start_hour'] = df['start_time'].dt.hour
    df['start_dayofweek'] = df['start_time'].dt.dayofweek # Monday=0, Sunday=6
    df['start_dayname'] = df['start_time'].dt.day_name()
    df['start_date'] = df['start_time'].dt.date
    print("Extracted hour, day of week, day name, and date features from 'start_time'.")

    # Calculate average speed (km/h)
    epsilon = 1e-6
    df['avg_speed_kmh'] = df['distance_km'] / ((df['trip_duration_minutes'] + epsilon) / 60)
    df.loc[df['distance_km'] == 0, 'avg_speed_kmh'] = 0
    df.loc[df['trip_duration_minutes'] <= 0, 'avg_speed_kmh'] = np.nan # Handle non-positive duration
    print("Calculated 'avg_speed_kmh'.")
else:
    print("Skipping time-based feature engineering due to timestamp conversion issues.")


print("\n--- Feature Engineering: Rolling Window & RFM-like Features ---")

# Define metrics and window size (use ride count window)
metrics_to_roll = ['fare', 'distance_km', 'trip_duration_minutes', 'cancelled', 'rating']
window_size = 3 # Use a small window like 3 previous rides due to data size

# --- User Rolling Features ---
print("Calculating user rolling features...")
df = df.sort_values(by=['user_id', 'start_time']) # Sort for user grouping

for metric in metrics_to_roll:
    if metric in df.columns:
        df[f'user_prev_{metric}'] = df.groupby('user_id')[metric].shift(1)
        df[f'user_rolling_mean_{window_size}_{metric}'] = df.groupby('user_id')[metric].shift(1).rolling(window=window_size, min_periods=1).mean()
        if metric not in ['cancelled', 'rating']: # Sum makes less sense for rate/rating
             df[f'user_rolling_sum_{window_size}_{metric}'] = df.groupby('user_id')[metric].shift(1).rolling(window=window_size, min_periods=1).sum()

df[f'user_rolling_cancel_rate_{window_size}'] = df.groupby('user_id')['cancelled'].shift(1).rolling(window=window_size, min_periods=1).mean()

# --- Driver Rolling Features ---
print("Calculating driver rolling features...")
df = df.sort_values(by=['driver_id', 'start_time']) # Sort for driver grouping

for metric in metrics_to_roll:
     if metric in df.columns:
        df[f'driver_prev_{metric}'] = df.groupby('driver_id')[metric].shift(1)
        df[f'driver_rolling_mean_{window_size}_{metric}'] = df.groupby('driver_id')[metric].shift(1).rolling(window=window_size, min_periods=1).mean()
        if metric not in ['cancelled', 'rating']:
            df[f'driver_rolling_sum_{window_size}_{metric}'] = df.groupby('driver_id')[metric].shift(1).rolling(window=window_size, min_periods=1).sum()

df[f'driver_rolling_cancel_rate_{window_size}'] = df.groupby('driver_id')['cancelled'].shift(1).rolling(window=window_size, min_periods=1).mean()

# --- RFM-like Features (User-based) ---
print("Calculating user RFM-like features...")
df = df.sort_values(by=['user_id', 'start_time']) # Ensure correct order

# Frequency
df['user_ride_count_before'] = df.groupby('user_id').cumcount()

# Recency
if time_conversion_success:
    df['user_prev_start_time'] = df.groupby('user_id')['start_time'].shift(1)
    df['user_days_since_last_ride'] = (df['start_time'] - df['user_prev_start_time']).dt.total_seconds() / (60*60*24)
    df = df.drop(columns=['user_prev_start_time'], errors='ignore')
else:
    df['user_days_since_last_ride'] = np.nan # Cannot calculate without valid timestamps

# Monetary (Cumulative & Average)
df['user_total_fare_before'] = df.groupby('user_id')['fare'].shift(1).expanding().sum()
df['user_avg_fare_before'] = df.groupby('user_id')['fare'].shift(1).expanding().mean()

# Behavioral (Cancellation Rate)
df['user_total_cancelled_before'] = df.groupby('user_id')['cancelled'].shift(1).expanding().sum()
df['user_cancel_rate_before'] = (df['user_total_cancelled_before'] / df['user_ride_count_before'])
df = df.drop(columns=['user_total_cancelled_before'], errors='ignore')

# --- Handle NaNs from Feature Engineering ---
print("Handling NaNs created during feature engineering...")
# Fill NaNs from shift/rolling/expanding (implying no prior history)
fill_zero_cols = [col for col in df.columns if 'prev_' in col or 'rolling_' in col or '_before' in col]
# Careful with recency - NaN means 'first ride'. Use a distinct value.
if 'user_days_since_last_ride' in fill_zero_cols:
    fill_zero_cols.remove('user_days_since_last_ride')

for col in fill_zero_cols:
    if col in df.columns:
        df[col] = df[col].fillna(0)

# Handle recency NaN (first ride) - fill with -1
if 'user_days_since_last_ride' in df.columns:
    df['user_days_since_last_ride'] = df['user_days_since_last_ride'].fillna(-1)

# Handle potential 0/0 division in user_cancel_rate_before (first ride) -> fill with 0
if 'user_cancel_rate_before' in df.columns:
    df['user_cancel_rate_before'] = df['user_cancel_rate_before'].fillna(0)

# Handle potential NaN in avg_speed_kmh if duration was non-positive
if 'avg_speed_kmh' in df.columns:
    df['avg_speed_kmh'] = df['avg_speed_kmh'].fillna(0) # Or another appropriate value


# Sort back to original time order
df = df.sort_values(by='start_time').reset_index(drop=True)

print("\n--- Sample of DataFrame with New Features ---")
print(df.head())
print(f"\nTotal features now: {df.shape[1]}")


# --- 5. Univariate Analysis & Visualization (Including Basic Engineered Features) ---

print("\n--- Univariate Analysis ---")

# Distribution of Numerical Features
numerical_features = ['fare', 'distance_km', 'rating', 'trip_duration_minutes', 'avg_speed_kmh']
print("Visualizing distributions of key numerical features...")
for col in numerical_features:
    if col in df.columns:
        plt.figure(figsize=(10, 4))
        plt.subplot(1, 2, 1)
        sns.histplot(df[col].dropna(), kde=True, bins=15) # dropna() just in case
        plt.title(f'Distribution of {col}')
        plt.subplot(1, 2, 2)
        sns.boxplot(y=df[col].dropna())
        plt.title(f'Box Plot of {col}')
        plt.tight_layout()
        plt.show()

# Frequency of Categorical Features
categorical_features = ['pickup_location', 'dropoff_location', 'cancelled', 'start_hour', 'start_dayname']
print("Visualizing frequencies of key categorical features...")
for col in categorical_features:
     if col in df.columns:
        plt.figure(figsize=(10, 5))
        order = None
        xtick_rotation = 0
        if df[col].nunique() > 10 and col not in ['start_hour', 'start_dayname']:
             # For high cardinality, show top N
             top_n = 10
             order = df[col].value_counts().nlargest(top_n).index
             plt.title(f'Frequency of Top {top_n} {col}')
             xtick_rotation = 45
        elif col == 'start_dayname':
            days_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
            order = [day for day in days_order if day in df[col].unique()]
            plt.title(f'Frequency of {col}')
        elif col == 'start_hour':
             order = sorted(df[col].unique())
             plt.title(f'Frequency of {col}')
        else:
            order = df[col].value_counts().index
            plt.title(f'Frequency of {col}')

        sns.countplot(data=df, x=col, order=order, palette='viridis')
        plt.xticks(rotation=xtick_rotation)
        plt.tight_layout()
        plt.show()

# --- 6. Multivariate Analysis & Visualization ---

print("\n--- Multivariate Analysis (Original & Basic Engineered Features) ---")

# Correlation between numerical features (excluding highly numerous engineered ones for now)
base_numeric_cols = ['fare', 'distance_km', 'rating', 'cancelled']
if 'trip_duration_minutes' in df.columns: base_numeric_cols.append('trip_duration_minutes')
if 'avg_speed_kmh' in df.columns: base_numeric_cols.append('avg_speed_kmh')

plt.figure(figsize=(10, 8))
correlation_matrix = df[base_numeric_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix (Base Numerical Features)')
plt.show()

# Fare vs. Distance
if 'distance_km' in df.columns and 'fare' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x='distance_km', y='fare', hue='cancelled', style='rating', palette='magma', s=100)
    plt.title('Fare vs. Distance (Colored by Cancellation, Styled by Rating)')
    plt.xlabel('Distance (km)')
    plt.ylabel('Fare')
    plt.legend(title='Cancelled | Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# Ride Volume Over Time (Hourly & Daily)
if 'start_hour' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.countplot(data=df, x='start_hour', hue='cancelled', palette='Set1', dodge=True, order=sorted(df['start_hour'].unique()))
    plt.title('Ride Volume by Hour of Day (Split by Cancellation)')
    plt.xlabel('Hour of Day (0-23)')
    plt.ylabel('Number of Rides')
    plt.legend(title='Cancelled (0=No, 1=Yes)')
    plt.tight_layout()
    plt.show()

if 'start_dayname' in df.columns:
    plt.figure(figsize=(10, 5))
    days_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    order = [day for day in days_order if day in df['start_dayname'].unique()]
    sns.countplot(data=df, x='start_dayname', order=order, hue='cancelled', palette='Set2', dodge=True)
    plt.title('Ride Volume by Day of Week (Split by Cancellation)')
    plt.xlabel('Day of Week')
    plt.ylabel('Number of Rides')
    plt.legend(title='Cancelled (0=No, 1=Yes)')
    plt.tight_layout()
    plt.show()

# Rating & Fare Distribution by Cancellation Status
if 'rating' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df, x='cancelled', y='rating', palette='pastel')
    plt.title('Rating Distribution for Cancelled vs. Completed Rides')
    plt.xlabel('Cancelled (0 = No, 1 = Yes)')
    plt.ylabel('Rating (1-5)')
    plt.xticks([0, 1], ['Completed', 'Cancelled'])
    plt.show()

if 'fare' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df, x='cancelled', y='fare', palette='pastel')
    plt.title('Fare Distribution for Cancelled vs. Completed Rides')
    plt.xlabel('Cancelled (0 = No, 1 = Yes)')
    plt.ylabel('Fare')
    plt.xticks([0, 1], ['Completed', 'Cancelled'])
    plt.show()

# Top Pickup and Dropoff Locations
n_top = 5
if 'pickup_location' in df.columns and 'dropoff_location' in df.columns:
    top_pickups = df['pickup_location'].value_counts().nlargest(n_top).index
    top_dropoffs = df['dropoff_location'].value_counts().nlargest(n_top).index

    plt.figure(figsize=(14, 5))
    plt.subplot(1, 2, 1)
    sns.countplot(data=df, y='pickup_location', order=top_pickups, palette='Blues_d')
    plt.title(f'Top {n_top} Pickup Locations')
    plt.xlabel('Number of Rides')
    plt.ylabel('Pickup Location')
    plt.subplot(1, 2, 2)
    sns.countplot(data=df, y='dropoff_location', order=top_dropoffs, palette='Greens_d')
    plt.title(f'Top {n_top} Dropoff Locations')
    plt.xlabel('Number of Rides')
    plt.ylabel('Dropoff Location')
    plt.tight_layout()
    plt.show()


print("\n--- Analysis of Engineered Features (Rolling Window & RFM) ---")

# --- Rolling Features vs. Cancellation ---
rolling_features_to_plot = [
    f'user_rolling_cancel_rate_{window_size}',
    f'driver_rolling_cancel_rate_{window_size}',
    f'user_rolling_mean_{window_size}_fare',
    f'driver_rolling_mean_{window_size}_trip_duration_minutes'
]
plt.figure(figsize=(15, 10))
plot_num = 1
print("Visualizing rolling features vs cancellation...")
for feature in rolling_features_to_plot:
    if feature in df.columns:
        plt.subplot(2, 2, plot_num)
        sns.boxplot(data=df, x='cancelled', y=feature, palette='vlag')
        plt.title(f'{feature} vs. Cancellation')
        plt.xlabel('Cancelled (0=No, 1=Yes)')
        plt.xticks([0, 1], ['Completed', 'Cancelled'])
        plot_num += 1
plt.suptitle('Distribution of Rolling Features by Cancellation Status', y=1.02)
plt.tight_layout()
plt.show()

# --- RFM-like Features vs. Cancellation ---
rfm_features_to_plot = [
    'user_ride_count_before',
    'user_days_since_last_ride',
    'user_avg_fare_before',
    'user_cancel_rate_before'
]
plt.figure(figsize=(15, 10))
plot_num = 1
print("Visualizing RFM-like features vs cancellation...")
for feature in rfm_features_to_plot:
     if feature in df.columns:
        plt.subplot(2, 2, plot_num)
        sns.violinplot(data=df, x='cancelled', y=feature, palette='coolwarm', inner='quartile')
        plt.title(f'{feature} vs. Cancellation')
        plt.xlabel('Cancelled (0=No, 1=Yes)')
        plt.xticks([0, 1], ['Completed', 'Cancelled'])
        if feature == 'user_days_since_last_ride':
            plt.ylabel('Days Since Last Ride (-1=First Ride)')
        plot_num += 1
plt.suptitle('Distribution of RFM-like Features by Cancellation Status', y=1.02)
plt.tight_layout()
plt.show()

# --- Relationship between RFM-like Features ---
if all(f in df.columns for f in ['user_ride_count_before', 'user_days_since_last_ride', 'user_avg_fare_before', 'user_cancel_rate_before']):
    print("Visualizing RFM feature relationships...")
    plt.figure(figsize=(10, 7))
    scatter = sns.scatterplot(
        data=df,
        x='user_ride_count_before',
        y='user_days_since_last_ride',
        hue='user_avg_fare_before',
        size='user_cancel_rate_before',
        palette='viridis',
        sizes=(20, 200),
        alpha=0.7
    )
    plt.title('User RFM-like Features: Recency vs. Frequency')
    plt.xlabel('User Ride Count (Frequency)')
    plt.ylabel('Days Since Last Ride (Recency, -1=First)')
    plt.legend(title='Avg Fare | Cancel Rate', bbox_to_anchor=(1.05, 1), loc='upper left')
    current_ticks = plt.yticks()[0]
    tick_labels = [str(int(t)) if t != -1 else 'First' for t in current_ticks]
    plt.yticks(current_ticks, tick_labels) # Adjust labels for recency
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.tight_layout()
    plt.show()

# --- Updated Correlation Matrix (Including All Numerical Features) ---
print("\n--- Updated Correlation Matrix (Including Engineered Features) ---")
numerical_cols_updated = df.select_dtypes(include=np.number).columns
# Limit features for readability if necessary
cols_for_heatmap = numerical_cols_updated # Or select a subset

# Increase figure size if many features
heatmap_size = max(15, len(cols_for_heatmap) * 0.4) # Dynamic sizing
plt.figure(figsize=(heatmap_size, heatmap_size * 0.8))
correlation_matrix_updated = df[cols_for_heatmap].corr()
sns.heatmap(correlation_matrix_updated, annot=False, cmap='coolwarm', fmt=".2f") # Keep annot=False for large matrices
plt.title('Correlation Matrix (Including All Numerical Features)')
plt.show()

# Specifically check correlation with the target variable 'cancelled'
print("\n--- Correlation with Target Variable ('cancelled') ---")
if 'cancelled' in correlation_matrix_updated:
    corr_with_target = correlation_matrix_updated['cancelled'].sort_values(ascending=False)
    # Filter out self-correlation and display significant correlations
    print(corr_with_target[(corr_with_target < 0.99) & (corr_with_target > -0.99)])
else:
    print("'cancelled' column not found in correlation matrix.")


# --- 7. EDA Summary ---
print("\n--- EDA Summary ---")
print(f"Dataset contains {df.shape[0]} rides and {df.shape[1]} features (including engineered ones).")
if time_conversion_success:
    print(f"Time range of data: {df['start_time'].min()} to {df['start_time'].max()}")
else:
     print("Time range not determined due to timestamp conversion issues.")

print("\nData Quality Checks:")
print(f"  - Missing Values: {'None Found' if missing_values.sum() == 0 else f'{missing_values.sum()} Found'}")
print(f"  - Duplicate Rows: {'None Found' if duplicate_rows == 0 else f'{duplicate_rows} Found'}")
print(f"  - Timestamps converted to datetime: {'Yes' if time_conversion_success else 'No'}")
if 'rating' in df.columns: print(f"  - Ratings are within expected range (1-5): {df['rating'].between(1, 5, inclusive='both').all()}")
if 'cancelled' in df.columns: print(f"  - 'cancelled' column contains only 0s and 1s: {all(val in [0, 1] for val in df['cancelled'].unique())}")

print("\nKey Observations (based on sample data):")
if time_conversion_success and 'trip_duration_minutes' in df.columns:
    print(f"- Average trip duration: {df['trip_duration_minutes'].mean():.2f} minutes (Note: Constant 30 mins in sample).")
if 'avg_speed_kmh' in df.columns: print(f"- Average speed: {df['avg_speed_kmh'].mean():.2f} km/h.")
if 'fare' in df.columns and 'distance_km' in df.columns and 'cancelled' in correlation_matrix:
    print(f"- Strong positive correlation observed between 'fare' and 'distance_km' ({correlation_matrix.loc['fare', 'distance_km']:.2f}).")
if 'cancelled' in df.columns: print(f"- Overall ride cancellation rate: {df['cancelled'].mean()*100:.1f}%")
if 'pickup_location' in df.columns: print(f"- Top pickup locations (sample): {list(top_pickups)}")
if 'dropoff_location' in df.columns: print(f"- Top dropoff locations (sample): {list(top_dropoffs)}")
if 'start_hour' in df.columns: print("- Ride volume varies by hour.") # Add specific peak hours if clear
if 'start_dayname' in df.columns: print("- Ride volume varies by day of the week.") # Add specific peak days if clear
print("- Engineered features (rolling windows, RFM) capture user/driver history.")
print("- Visualizations suggest potential differences in recent behavior (e.g., rolling cancel rates) and long-term user characteristics (e.g., frequency, recency) between cancelled and completed rides.")
print("- Correlation analysis with 'cancelled' highlights features potentially predictive of cancellation (e.g., [mention 1-2 features with highest absolute correlation from the output]).")

print("\nNext Steps: Proceed to User Behavior Analysis, Predictive Modeling, and Operational Insights.")
```

# 2. User Behavior Analysis

```python
# (Assuming the code from the previous steps, including loading 'df' and initial checks, has been run)

# --- Part 2: User Behavior Analysis ---

print("\n\n" + "="*50)
print("Part 2: User Behavior Analysis")
print("="*50 + "\n")

# --- Pre-computation and Checks --- (Keep the checks from the previous version)
# Required columns for this section
required_cols = ['ride_id', 'user_id', 'driver_id', 'start_time', 'end_time',
                 'pickup_location', 'dropoff_location', 'fare', 'distance_km',
                 'rating', 'cancelled']

if 'df' not in locals() or not isinstance(df, pd.DataFrame):
    print("Error: DataFrame 'df' not found. Please run EDA first.")
    exit()
if not all(col in df.columns for col in required_cols):
    print(f"Error: DataFrame 'df' is missing one or more required columns: {required_cols}. Please check EDA.")
    exit()

try:
    df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
    df['cancelled'] = pd.to_numeric(df['cancelled'], errors='coerce')
    print("Converted 'rating' and 'cancelled' to numeric.")
except Exception as e:
    print(f"Warning: Could not convert rating/cancelled to numeric - {e}")

if not pd.api.types.is_datetime64_any_dtype(df['start_time']):
     try:
         df['start_time'] = pd.to_datetime(df['start_time'])
         print("Converted 'start_time' to datetime.")
     except Exception as e:
         print(f"Error: Could not convert 'start_time' to datetime - {e}. Churn analysis might be affected.")
         # exit()


# --- 2.1 User Segmentation by Ride Frequency and Spending ---

print("\n--- 2.1 User Segmentation (Frequency & Spending) ---")

if pd.api.types.is_datetime64_any_dtype(df['start_time']):
    max_time = df['start_time'].max()
    print(f"Dataset maximum timestamp for recency: {max_time}")
else:
    max_time = None
    print("Warning: 'start_time' is not datetime. Recency calculations will be skipped or inaccurate.")

agg_dict = {
    'total_rides': ('ride_id', 'count'),
    'total_spending': ('fare', 'sum'),
    'avg_spending': ('fare', 'mean'),
    'first_ride_time': ('start_time', 'min'),
    'last_ride_time': ('start_time', 'max'),
    'total_cancelled': ('cancelled', lambda x: (x.dropna() == 1).sum()),
    'total_low_ratings': ('rating', lambda x: (x.dropna() <= 2).sum()),
    'avg_rating': ('rating', 'mean')
}

# Adjust agg_dict based on available columns/types (same as before)
if not pd.api.types.is_datetime64_any_dtype(df['start_time']):
    agg_dict.pop('first_ride_time', None); agg_dict.pop('last_ride_time', None)
if 'cancelled' not in df.columns: agg_dict.pop('total_cancelled', None)
if 'rating' not in df.columns: agg_dict.pop('total_low_ratings', None); agg_dict.pop('avg_rating', None)


try:
    user_summary = df.groupby('user_id').agg(**agg_dict).reset_index()

    # --- MODIFIED RECENTY CALCULATION ---
    if max_time and 'last_ride_time' in user_summary.columns:
        # Calculate timedelta
        time_diff = max_time - user_summary['last_ride_time']
        # Convert timedelta to fractional days (total seconds / seconds in a day)
        user_summary['recency_days_fractional'] = time_diff.dt.total_seconds() / (60 * 60)
        print("Calculated 'recency_days_fractional' (non-rounded days).")
        # Keep last ride date for churn check
        user_summary['last_ride_date'] = user_summary['last_ride_time'].dt.date
    else:
        user_summary['recency_days_fractional'] = np.nan
        user_summary['last_ride_date'] = pd.NaT
        print("Skipping recency calculation.")
    # --- END OF MODIFICATION ---


    # Calculate user-level rates (same as before)
    if 'total_cancelled' in user_summary.columns:
        user_summary['cancellation_rate'] = (user_summary['total_cancelled'] / user_summary['total_rides']).replace([np.inf, -np.inf], 0).fillna(0)
    else: user_summary['cancellation_rate'] = np.nan
    if 'total_low_ratings' in user_summary.columns:
        low_rating_threshold = 2
        user_summary['low_rating_rate'] = (user_summary['total_low_ratings'] / user_summary['total_rides']).replace([np.inf, -np.inf], 0).fillna(0)
    else: user_summary['low_rating_rate'] = np.nan


    # --- Define Segments using Quantiles --- (Keep the robust segmentation logic from before)
    freq_labels = ['Low Freq', 'Med-Low Freq', 'Med-High Freq', 'High Freq']
    spend_labels = ['Low Spend', 'Med-Low Spend', 'Med-High Spend', 'High Spend']
    segmentation_possible = True
    try:
        if user_summary.shape[0] >= 4: user_summary['freq_segment'] = pd.qcut(user_summary['total_rides'], q=4, labels=freq_labels, duplicates='drop')
        else: user_summary['freq_segment'] = 'Single Freq Group'; segmentation_possible = False
    except ValueError:
        try: user_summary['freq_segment'] = pd.qcut(user_summary['total_rides'].rank(method='first'), q=4, labels=freq_labels, duplicates='drop')
        except: user_summary['freq_segment'] = 'Segmentation Failed'; segmentation_possible = False
    try:
        if user_summary.shape[0] >= 4: user_summary['spend_segment'] = pd.qcut(user_summary['total_spending'], q=4, labels=spend_labels, duplicates='drop')
        else: user_summary['spend_segment'] = 'Single Spend Group'; segmentation_possible = False
    except ValueError:
        try: user_summary['spend_segment'] = pd.qcut(user_summary['total_spending'].rank(method='first'), q=4, labels=spend_labels, duplicates='drop')
        except: user_summary['spend_segment'] = 'Segmentation Failed'; segmentation_possible = False

    if segmentation_possible and 'freq_segment' in user_summary.columns and 'spend_segment' in user_summary.columns:
        user_summary['user_segment'] = user_summary['freq_segment'].astype(str) + ' - ' + user_summary['spend_segment'].astype(str)
    else: user_summary['user_segment'] = 'N/A'


    print("\nUser Summary DataFrame with Segments:")
    # Display the new fractional recency column
    print(user_summary[['user_id', 'total_rides', 'total_spending', 'last_ride_time', 'recency_days_fractional', 'last_ride_date', 'user_segment']].head())

    if 'user_segment' in user_summary.columns and user_summary['user_segment'].nunique() > 1 and user_summary['user_segment'].iloc[0] != 'N/A':
        print("\nSegment Counts:")
        print(user_summary['user_segment'].value_counts())

        # --- Visualize Segments ---
        plt.figure(figsize=(10, 7))
        sns.scatterplot(data=user_summary, x='total_rides', y='total_spending', hue='user_segment',
                        size='recency_days_fractional', # Size by fractional recency
                        sizes=(50, 250), alpha=0.8, palette='tab10')
        plt.title('User Segmentation: Frequency vs. Total Spending')
        plt.xlabel('Total Rides (Frequency)')
        plt.ylabel('Total Spending (Monetary)')
        plt.legend(title='Segment | Recency (Fractional Days)', bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.tight_layout()
        plt.show()
    else:
        print("Skipping segment visualization as unique segments could not be created.")

except Exception as e:
    print(f"An error occurred during user summary aggregation or segmentation: {e}")
    user_summary = pd.DataFrame()


# --- 2.2 Analyze Trends in Cancellations and Low Ratings ---
if not user_summary.empty:
    print("\n--- 2.2 Analysis of Cancellations and Low Ratings ---")

    overall_cancel_rate = df['cancelled'].mean() if 'cancelled' in df.columns else np.nan
    overall_low_rating_rate = (df['rating'] <= low_rating_threshold).mean() if 'rating' in df.columns else np.nan
    print(f"Overall Cancellation Rate: {overall_cancel_rate:.2%}" if pd.notna(overall_cancel_rate) else "Cancellation rate N/A")
    print(f"Overall Low Rating (<= {low_rating_threshold}) Rate: {overall_low_rating_rate:.2%}" if pd.notna(overall_low_rating_rate) else "Low rating rate N/A")

    # --- Rates per User Segment ---
    if 'user_segment' in user_summary.columns and user_summary['user_segment'].nunique() > 1 and user_summary['user_segment'].iloc[0] != 'N/A':
        segment_analysis = user_summary.groupby('user_segment').agg(
            avg_cancel_rate=('cancellation_rate', 'mean'),
            avg_low_rating_rate=('low_rating_rate', 'mean'),
            avg_recency_frac=('recency_days_fractional', 'mean'), # Use fractional recency here
            user_count=('user_id', 'count')
        ).reset_index()
        print("\nAverage Rates and Recency by User Segment:")
        print(segment_analysis.sort_values(by='avg_cancel_rate', ascending=False))

        # Visualizations (Bar Plot and Box Plot - same as before)
        # Bar Plot
        plt.figure(figsize=(14, 6))
        plt.subplot(1, 2, 1); sns.barplot(data=segment_analysis, x='avg_cancel_rate', y='user_segment', palette='Reds_d', order=segment_analysis.sort_values('avg_cancel_rate', ascending=False)['user_segment']); plt.title('Average Cancellation Rate by User Segment'); plt.xlabel('Average Cancellation Rate'); plt.ylabel('User Segment')
        plt.subplot(1, 2, 2); sns.barplot(data=segment_analysis, x='avg_low_rating_rate', y='user_segment', palette='Blues_d', order=segment_analysis.sort_values('avg_low_rating_rate', ascending=False)['user_segment']); plt.title(f'Average Low Rating (<= {low_rating_threshold}) Rate by Segment'); plt.xlabel(f'Average Low Rating Rate'); plt.ylabel('')
        plt.tight_layout(); plt.show()
        # Box Plot
        plt.figure(figsize=(16, 7))
        plt.subplot(1, 2, 1); order_cancel = user_summary.groupby('user_segment')['cancellation_rate'].median().sort_values().index; sns.boxplot(data=user_summary, x='cancellation_rate', y='user_segment', order=order_cancel, palette='Reds_d'); plt.title('Distribution of Cancellation Rates by User Segment'); plt.xlabel('User Cancellation Rate'); plt.ylabel('User Segment'); plt.grid(True, axis='x', linestyle='--', alpha=0.6)
        plt.subplot(1, 2, 2); order_rating = user_summary.groupby('user_segment')['low_rating_rate'].median().sort_values().index; sns.boxplot(data=user_summary, x='low_rating_rate', y='user_segment', order=order_rating, palette='Blues_d'); plt.title(f'Distribution of Low Rating (<= {low_rating_threshold}) Rates by Segment'); plt.xlabel(f'User Low Rating Rate'); plt.ylabel(''); plt.grid(True, axis='x', linestyle='--', alpha=0.6)
        plt.tight_layout(); plt.show()
    else:
        print("Skipping segment-based rate analysis.")

    # --- Correlation between user metrics ---
    print("\nCorrelation of User Summary Metrics:")
    # Use the new fractional recency column for correlation
    user_metrics_for_corr = ['total_rides', 'total_spending', 'avg_spending', 'recency_days_fractional', 'cancellation_rate', 'low_rating_rate', 'avg_rating']
    valid_user_metrics = [col for col in user_metrics_for_corr if col in user_summary.columns and user_summary[col].notna().any()]
    if len(valid_user_metrics) > 1:
        user_metrics_corr = user_summary[valid_user_metrics].corr()
        plt.figure(figsize=(8, 6)); sns.heatmap(user_metrics_corr, annot=True, cmap='coolwarm', fmt=".2f"); plt.title('Correlation Matrix of User Summary Metrics'); plt.show()
        if 'cancellation_rate' in user_metrics_corr: print("\nCorrelations with Cancellation Rate:"); print(user_metrics_corr['cancellation_rate'].sort_values(ascending=False))
        if 'low_rating_rate' in user_metrics_corr: print(f"\nCorrelations with Low Rating (<= {low_rating_threshold}) Rate:"); print(user_metrics_corr['low_rating_rate'].sort_values(ascending=False))
    else: print("Skipping user metrics correlation matrix.")

# --- 2.3 Identify Churn Indicators ---
if not user_summary.empty and 'last_ride_date' in user_summary.columns and pd.api.types.is_datetime64_any_dtype(df['start_time']):
    print("\n--- 2.3 Churn Indicator Analysis ---")
    print("NOTE: 'Churn' is based on inactivity *within this short dataset's timeframe*.")

    last_date_in_data = df['start_time'].dt.date.max()
    print(f"Last calendar date in dataset: {last_date_in_data}")

    # Define potential churn based on last ride DATE comparison (Correctly identifies U1)
    user_summary['potential_churn'] = user_summary['last_ride_date'] < last_date_in_data
    print("Defined 'potential_churn' based on whether last ride was before the last calendar date in data.")

    churned_count = user_summary['potential_churn'].sum()
    total_users = user_summary.shape[0]
    print(f"Number of potentially churned users (last ride before {last_date_in_data}): {churned_count} out of {total_users}")
    print("Users identified as potentially churned:")
    print(user_summary[user_summary['potential_churn']][['user_id', 'last_ride_date', 'recency_days_fractional']]) # Show fractional recency too

    # --- Compare Characteristics ---
    if user_summary['potential_churn'].nunique() > 1:
        churn_comparison = user_summary.groupby('potential_churn').agg(
            avg_total_rides=('total_rides', 'mean'),
            avg_total_spending=('total_spending', 'mean'),
            avg_recency_frac=('recency_days_fractional', 'mean'), # Use fractional recency
            avg_cancel_rate=('cancellation_rate', 'mean'),
            avg_low_rating_rate=('low_rating_rate', 'mean'),
            avg_rating=('avg_rating', 'mean'),
            user_count=('user_id', 'count')
        ).reset_index()
        churn_comparison['potential_churn'] = churn_comparison['potential_churn'].map({True: 'Churned (Inactive on Last Day)', False: 'Active (Active on Last Day)'})
        print("\nComparison of Active vs. Potentially Churned Users:")
        print(churn_comparison)
    else: print("\nCannot compare churned vs active users - all users fall into one category.")

    # Visualize Recency Distribution (using the fractional days)
    plt.figure(figsize=(8, 5))
    sns.histplot(user_summary['recency_days_fractional'].dropna(), bins=10, kde=True) # Use more bins for fractional data
    plt.title('Distribution of User Recency (Fractional Days Since Last Ride)')
    plt.xlabel('Recency (Fractional Days)')
    plt.ylabel('Number of Users')
    plt.grid(True, axis='y', linestyle='--', alpha=0.7)
    plt.show()
    print(f"Note on Recency Plot: Shows distribution of fractional days since last ride relative to {max_time}.")

else:
    print("\nSkipping Churn Indicator Analysis due to missing data or previous errors.")


# --- User Behavior Analysis Summary --- (Update to mention fractional recency)
print("\n--- User Behavior Analysis Summary ---")
if not user_summary.empty:
    if 'user_segment' in user_summary.columns and user_summary['user_segment'].iloc[0] != 'N/A': print("- Users segmented based on ride frequency and total spending.") # ... (rest of summary similar to before)
    else: print("- User segmentation could not be performed or resulted in non-unique segments.")
    if pd.notna(overall_cancel_rate): print(f"- Overall cancellation rate: {overall_cancel_rate:.2%}.")
    if pd.notna(overall_low_rating_rate): print(f"- Overall low rating (<= {low_rating_threshold}) rate: {overall_low_rating_rate:.2%}.")
    if 'segment_analysis' in locals(): print("- Analysis by segment reveals potential differences in average rates and fractional recency across groups.")
    if 'user_metrics_corr' in locals(): print("- Correlation analysis highlights linear relationships between user metrics (including fractional recency).")
    if 'potential_churn' in user_summary.columns:
        print(f"- Potential churn indicators identified: {churned_count} users whose last ride was before {last_date_in_data}.")
        if 'churn_comparison' in locals(): print("- Comparison shows potential differences in behavior between active and potentially churned users (interpret cautiously).")
        else: print("- Churn comparison skipped as all users were in one category (active/churned).")
    else: print("- Churn analysis skipped.")
else: print("- User Behavior Analysis skipped due to errors in data preparation or aggregation.")
```

# 3. Predictive Modeling

```python
# Ensure previous imports are present
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.metrics import precision_score, recall_score, f1_score, make_scorer, classification_report, confusion_matrix, accuracy_score
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# --- Part 3: Predictive Modeling (Revised with SMOTE k=2 in CV) ---

print("\n\n" + "="*50)
print("Part 3: Predictive Modeling (Using SMOTE k=2 in Stratified K-Fold CV)")
print("="*50 + "\n")

# Assume 'df' exists from previous steps
if 'df' not in locals() or not isinstance(df, pd.DataFrame):
    print("Error: DataFrame 'df' not found.")
    exit()

# --- 3.1 Feature Selection and Preparation (Same as before) ---
print("--- 3.1 Feature Selection ---")
target = 'cancelled'
potential_features = [col for col in df.columns if col not in [
    'ride_id', 'user_id', 'driver_id', 'cancelled',
    'start_time', 'end_time', 'start_date',
    'rating', 'trip_duration_minutes', 'avg_speed_kmh',
    'user_prev_start_time', 'start_dayname'
]]

df_features = df[potential_features].copy()
df_features.replace([np.inf, -np.inf], np.nan, inplace=True)
X = df_features
y = df[target].copy()

print(f"Selected {X.shape[1]} potential features.")
categorical_features = X.select_dtypes(include=['object', 'category']).columns.tolist()
numerical_features = X.select_dtypes(include=np.number).columns.tolist()
print(f"\nCategorical features: {categorical_features}")
print(f"Numerical features: {numerical_features}")

# --- 3.2 Preprocessing Pipeline Definition (Same as before) ---
print("\n--- 3.2 Preprocessing Pipeline Setup ---")
numerical_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())])
categorical_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent')), ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
preprocessor = ColumnTransformer(transformers=[('num', numerical_transformer, numerical_features), ('cat', categorical_transformer, categorical_features)], remainder='passthrough')

# --- 3.3 Model Definition and Pipeline with SMOTE (k=2) ---
print("\n--- 3.3 Model Definition and Pipeline with SMOTE (k=2) ---")
model = RandomForestClassifier(random_state=42, n_estimators=100) # Removed class_weight

# --- *** ADJUSTED k_neighbors *** ---
n_minority = y.value_counts().min()
# Set k_neighbors to be less than the minimum expected in a training fold (which is 3)
smote_k_neighbors = min(n_minority - 2, 2) # Aim for 2, but ensure it's <= n_minority-2
smote_k_neighbors = max(1, smote_k_neighbors) # Ensure k is at least 1
print(f"Minority samples: {n_minority}. Setting SMOTE k_neighbors={smote_k_neighbors}")
smote = SMOTE(random_state=42, k_neighbors=smote_k_neighbors)
# --- **************************** ---

pipeline_with_smote = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('smote', smote),
    ('classifier', model)
])
print(f"Using model: {model.__class__.__name__} with SMOTE (k={smote_k_neighbors}) in pipeline")

# --- 3.4 Stratified K-Fold Cross-Validation ---
print("\n--- 3.4 Stratified K-Fold Cross-Validation (with SMOTE k=2) ---")

n_splits = min(n_minority, 4)
if n_splits < 2:
     print(f"Error: Smallest class ({n_minority} samples) too small for CV.")
     exit()

print(f"Using n_splits = {n_splits}")
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

scoring = {
    'accuracy': 'accuracy', 'precision_macro': 'precision_macro', 'recall_macro': 'recall_macro',
    'f1_macro': 'f1_macro', 'precision_class1': make_scorer(precision_score, pos_label=1, zero_division=0),
    'recall_class1': make_scorer(recall_score, pos_label=1, zero_division=0),
    'f1_class1': make_scorer(f1_score, pos_label=1, zero_division=0)
}

print(f"Running cross-validation with SMOTE (k={smote_k_neighbors})...")
try:
    cv_results = cross_validate(pipeline_with_smote, X, y, cv=skf, scoring=scoring, n_jobs=-1, error_score='raise')
    print("Cross-validation complete.")
    cv_successful = True

    avg_accuracy = np.mean(cv_results['test_accuracy'])
    avg_precision_macro = np.mean(cv_results['test_precision_macro'])
    avg_recall_macro = np.mean(cv_results['test_recall_macro'])
    avg_f1_macro = np.mean(cv_results['test_f1_macro'])
    avg_precision_class1 = np.mean(cv_results['test_precision_class1'])
    avg_recall_class1 = np.mean(cv_results['test_recall_class1'])
    avg_f1_class1 = np.mean(cv_results['test_f1_class1'])

except ValueError as e:
    print(f"\nError during cross-validation (with SMOTE k={smote_k_neighbors}): {e}")
    print("This might happen if SMOTE's k_neighbors is >= number of minority samples in a specific fold.")
    print("Even with k=2, if a fold has < 2 minority samples, it can fail. Data may be insufficient.")
    cv_successful = False
except Exception as e:
    print(f"\nAn unexpected error occurred during cross-validation (with SMOTE): {e}")
    cv_successful = False


# --- 3.5 Evaluation Results from Cross-Validation ---
if cv_successful:
    print("\n--- 3.5 Cross-Validation Evaluation Results (with SMOTE k=2) ---")
    print(f"Average Accuracy: {avg_accuracy:.4f}")
    print(f"Average Precision (Macro): {avg_precision_macro:.4f}")
    print(f"Average Recall (Macro): {avg_recall_macro:.4f}")
    print(f"Average F1-Score (Macro): {avg_f1_macro:.4f}")
    print("-" * 20)
    print("Metrics specific to Class 1 (Cancelled):")
    print(f"Average Precision (Class 1): {avg_precision_class1:.4f}")
    print(f"Average Recall (Class 1): {avg_recall_class1:.4f}")
    print(f"Average F1-Score (Class 1): {avg_f1_class1:.4f}")

    print("\nIndividual Fold Scores (for reference):")
    cv_results_df = pd.DataFrame(cv_results)
    print(cv_results_df)

    if cv_results_df['test_recall_class1'].sum() > 0:
        print("\nNote: SMOTE (k=2) resulted in some successful Class 1 predictions in at least one fold.")
    else:
        print("\nNote: SMOTE (k=2) did not result in any successful Class 1 predictions across folds.")

    # --- Optional: Train Final Model on All Data & Show Feature Importance ---
    print("\n--- Training Final Model on All Data (with SMOTE k=2 pipeline, for Feature Importance) ---")
    try:
        pipeline_with_smote.fit(X, y)
        print("Final model training complete.")
        if hasattr(model, 'feature_importances_'):
            print("\n--- Feature Importance (from model trained on all data with SMOTE k=2) ---")
            try:
                num_features_out = numerical_features
                cat_features_out = pipeline_with_smote.named_steps['preprocessor'].named_transformers_['cat'].named_steps['onehot'].get_feature_names_out(categorical_features)
                all_feature_names = num_features_out + list(cat_features_out)
                importances = pipeline_with_smote.named_steps['classifier'].feature_importances_
                if len(all_feature_names) == len(importances):
                    feature_importance_df = pd.DataFrame({'Feature': all_feature_names, 'Importance': importances})
                    feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False).head(15)
                    plt.figure(figsize=(10, 8)); sns.barplot(x='Importance', y='Feature', data=feature_importance_df, palette='viridis'); plt.title(f'Top 15 Feature Importances (SMOTE k={smote_k_neighbors}, Trained on All Data)'); plt.tight_layout(); plt.show()
                else: print(f"Mismatch in feature names ({len(all_feature_names)}) and importances ({len(importances)}). Skipping plot.")
            except Exception as e: print(f"Could not plot feature importance: {e}")
    except Exception as e: print(f"Error training final model on all data: {e}")

else:
    print("\nSkipping evaluation results as cross-validation failed.")


print("\nPredictive Modeling Summary:")
print("- Prepared data, handled infinities, set up preprocessing pipeline.")
print(f"- Used Stratified {n_splits}-Fold Cross-Validation with SMOTE (k={smote_k_neighbors}) to address severe class imbalance.")
if cv_successful:
    print(f"- Average CV Metrics: Accuracy={avg_accuracy:.4f}, Precision (Class 1)={avg_precision_class1:.4f}, Recall (Class 1)={avg_recall_class1:.4f}, F1 (Class 1)={avg_f1_class1:.4f}")
    if cv_results_df['test_recall_class1'].sum() == 0:
        print("- WARNING: Despite SMOTE (k=2), the model still failed to correctly predict any cancellations during cross-validation.")
    print("- Feature importances calculated from a model trained on the full dataset (using SMOTE pipeline).")
    print("- Conclusion: The dataset is severely limited for predicting cancellations. Even with SMOTE, the lack of sufficient real positive examples prevents reliable model training. More data is critical.")
else:
    print("- Cross-validation with SMOTE failed, likely due to insufficient minority samples within folds even for k=2.")
    print("- Conclusion: The dataset is too small for reliable modeling.")
```
